```python
#PIE CHART:
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
```

```python
df = pd.DataFrame({
    "etiquetas":['Manzanas','Bananas','Naranjas','Uvas'],
    "cantidad":[30,40,50,80]
})
df
```

```python
#Agregar etquetas LABELS
plt.pie(df['cantidad'],labels=df['etiquetas'],
       autopct='%1.1f%%');
```

```python
#Parmtro COLOR:
plt.pie(df['cantidad'],labels=df['etiquetas'],
       autopct='%1.1f%%',
        colors=['#87CEEB','#6495ED','#000080','#40E0D0']);
```

```python
#Parmetro SHADOW = Proyectar una Sombra:
plt.pie(df['cantidad'],labels=df['etiquetas'],
       autopct='%1.1f%%',
        colors=['#87CEEB','#6495ED','#000080','#40E0D0'],
       shadow = True);
```

```python
#Mtodo EXPLODE = Para separa las porciones de la torta:
plt.pie(df['cantidad'],labels=df['etiquetas'],
       autopct='%1.1f%%',
        colors=['#87CEEB','#6495ED','#000080','#40E0D0'],
       shadow = True,
       explode = (0,0,0,0.5));
```

```python
#Colocar el ttulo:
plt.pie(df['cantidad'],labels=df['etiquetas'],
       autopct='%1.1f%%',
        colors=['#87CEEB','#6495ED','#000080','#40E0D0'],
       shadow = True,
       explode = (0,0,0,0))
plt.title("El Pastel de Frutas de Ella");
```
