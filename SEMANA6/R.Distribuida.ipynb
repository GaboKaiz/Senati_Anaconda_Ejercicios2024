```python
#Representaci√≥n distribuida DISPLOT()
import seaborn as sns
```

```python
pinguinos = sns.load_dataset("penguins")
pinguinos.head()
```

```python
sns.displot(data=pinguinos,
           x = "body_mass_g",
            #KDE: Representaci√≥n de la densidad de la probabilidad estimada (Picos)
            kde = True,
            #COL: Comparaci√≥n con otro atributo ? ISLAS
            col = "island",
           );
```

```python
#RUGPLOT:
sns.displot(data=pinguinos,
           x = "flipper_length_mm",
            #KDE: Representaci√≥n de la densidad de la probabilidad estimada (Picos)
            kde = True
           );
```

```python
sns.rugplot(data=pinguinos, x="flipper_length_mm")
```
