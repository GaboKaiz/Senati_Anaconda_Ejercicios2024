```python
#Relaciones Estadísticas con RELPLOT():
import seaborn as sns
```

```python
propinas = sns.load_dataset("tips")
propinas
```

```python
sns.relplot(data = propinas,
           x = "total_bill",
           y = "tip");
```

```python
sns.relplot(data = propinas,
           x = "total_bill",
           y = "tip",
           kind = "line");
```

```python
sns.relplot(data = propinas,
           x = "total_bill",
           y = "tip",
           size = "size",
           hue = "smoker",
           style = "smoker",
           legend = None,
           col = "time");
```
