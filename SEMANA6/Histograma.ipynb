```python
#Crear un ejemplo con números aleatorios:
numeros = np.random.randn(1000)
#Mostramos:
numeros
#Lo pasamos los 1000 numros a un histograma:
plt.hist(numeros);
```

```python
#Agregamos un parámetrp BINS = Número de Barras:
plt.hist(numeros,bins=40);
```

```python
#Parámetro ALPHA = Establecer Transparencia:
plt.hist(numeros,bins=40, alpha=0.5);
```

```python
#Parámetro Color = Determinar el tinte o color (RGB o Hexadecimal):
plt.hist(numeros,bins=40, alpha=0.5, color="orange");
```

```python
#Parámetro EDGECOLOR= Color del borde de las columnas:
plt.hist(numeros,bins=40, alpha=0.5, color="orange", edgecolor='green');
```

```python
#Parámetro Histtype = Estilos visuales predefinidos:
plt.hist(numeros,bins=40, alpha=0.5, color="orange", edgecolor='green', histtype='step');
```

```python
#Histogramas superpuestos:
x1 = np.random.randn(1000)
x2 = np.random.randn(500)
x3 =np.random.randn(300)
plt.hist(x1, histtype='stepfilled',bins=40, alpha=0.3);
plt.hist(x2, histtype='stepfilled',bins=40, alpha=0.3);
plt.hist(x3, histtype='stepfilled',bins=40, alpha=0.3);
```
