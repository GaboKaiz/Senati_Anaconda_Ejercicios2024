```python
#Importamos matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
```

```python
#Ejercicio NÂ° 7:
df = pd.DataFrame({
    "Frutas":['Manzanas','Bananas','Cerezas','Duraznos'],
    "Cantidades":[40,25,15,20]
})
df
```

```python
plt.pie(df['Cantidades']);
```

```python
#Ejercicio NÂ° 8:

#Agregar etquetas LABELS
plt.pie(df['Cantidades'],labels=df['Frutas'],
       autopct='%1.1f%%');
```

```python
#Ejercicio NÂ° 9:
#Asignamos colores al pastel segun su fruta:
plt.pie(df['Cantidades'],labels=df['Frutas'],
       autopct='%1.1f%%',
        #colors=['#DC143C','#FFD700','#DB7093','#FF7F50'],
        colors=['red','yellow','darkred','orange'],
       shadow = True);
```

```python
#Agrega una sombra:
#ParÃ¡metro SHADOW = Proyectar una Sombra:
plt.pie(df['Cantidades'],labels=df['Frutas'],
       autopct='%1.1f%%',
        colors=['red','yellow','darkred','orange'],
       shadow = True);
```

```python
#Separa la porciÃ³n de las Cerezas del resto de las frutas en el grÃ¡fico:
```

```python
#MÃ©todo EXPLODE = Para separa las porciones de la torta:
plt.pie(df['Cantidades'],labels=df['Frutas'],
       autopct='%1.1f%%',
        colors=['red','yellow','darkred','orange'],
       shadow = True,
       explode = (0,0,0.5,0));
```

```python
#Finalmente, aÃ±ade al grÃ¡fico, el siguiente tÃ­tulo: 'DistribuciÃ³n de frutas
#en el almacÃ©n:
#Colocar el tÃ­tulo:
plt.pie(df['Cantidades'],labels=df['Frutas'],
       autopct='%1.1f%%',
        #colors=['#DC143C','#FFD700','#DB7093','#FF7F50'],
        colors=['red','yellow','darkred','orange'],
       shadow = True)
plt.title("DistribuciÃ³n de frutas en el almacÃ©n");
```
