```python
#Importamos matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
```

```python
#Ejercicio N° 7:
df = pd.DataFrame({
    "Frutas":['Manzanas','Bananas','Cerezas','Duraznos'],
    "Cantidades":[40,25,15,20]
})
df
```

```python
plt.pie(df['Cantidades']);
```

```python
#Ejercicio N° 8:

#Agregar etquetas LABELS
plt.pie(df['Cantidades'],labels=df['Frutas'],
       autopct='%1.1f%%');
```

```python
#Ejercicio N° 9:
#Asignamos colores al pastel segun su fruta:
plt.pie(df['Cantidades'],labels=df['Frutas'],
       autopct='%1.1f%%',
        #colors=['#DC143C','#FFD700','#DB7093','#FF7F50'],
        colors=['red','yellow','darkred','orange'],
       shadow = True);
```

```python
#Agrega una sombra:
#Parámetro SHADOW = Proyectar una Sombra:
plt.pie(df['Cantidades'],labels=df['Frutas'],
       autopct='%1.1f%%',
        colors=['red','yellow','darkred','orange'],
       shadow = True);
```

```python
#Separa la porción de las Cerezas del resto de las frutas en el gráfico:
```

```python
#Método EXPLODE = Para separa las porciones de la torta:
plt.pie(df['Cantidades'],labels=df['Frutas'],
       autopct='%1.1f%%',
        colors=['red','yellow','darkred','orange'],
       shadow = True,
       explode = (0,0,0.5,0));
```

```python
#Finalmente, añade al gráfico, el siguiente título: 'Distribución de frutas
#en el almacén:
#Colocar el título:
plt.pie(df['Cantidades'],labels=df['Frutas'],
       autopct='%1.1f%%',
        #colors=['#DC143C','#FFD700','#DB7093','#FF7F50'],
        colors=['red','yellow','darkred','orange'],
       shadow = True)
plt.title("Distribución de frutas en el almacén");
```
